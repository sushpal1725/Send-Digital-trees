```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt 
%matplotlib inline

all_registrations = pd.read_csv('all_registartions.csv', delimiter = ' ', names = ['day', 'my_date', 'my_time', 'user_id', 'event_type', 'birth_year', 'phone_type', 'location','source'])

daily_registrations = all_registrations.groupby('my_date').count()
daily_registrations.plot()

dailyregistrations = daily_registrations.reset_index(drop = True).event_type

x = dailyregistrations .index
y = dailyregistrations .values

coefs = np.polyfit(x, y, 1)
predict = np.poly1d(coefs)

from sklearn.metrics import r2_score
r2_score(y, predict(x))

x_test = np.linspace(0, 186)
y_pred = predict(x_test[:, None])
plt.scatter(x, y)
plt.plot(x_test, y_pred, c = 'r')
plt.show()

print(predict(186 + 28))

coefs = np.polyfit(x, y, 2)
predict = np.poly1d(coefs)

x_test = np.linspace(0, 186)
y_pred = predict(x_test[:, None])
plt.scatter(x, y)
plt.plot(x_test, y_pred, c = 'r')
plt.show()

print(predict(186 + 28))
    
coefs = np.polyfit(x, y, 3)
predict = np.poly1d(coefs)

x_test = np.linspace(0, 186)
y_pred = predict(x_test[:, None])
plt.scatter(x, y)
plt.plot(x_test, y_pred, c = 'r')
plt.show()

print(predict(186 + 28))

coefs = np.polyfit(x, y, 4)
predict = np.poly1d(coefs)

x_test = np.linspace(0, 186)
y_pred = predict(x_test[:, None])
plt.scatter(x, y)
plt.plot(x_test, y_pred, c = 'r')
plt.show()

print(predict(186 + 28))

coefs = np.polyfit(x, y, 5)
predict = np.poly1d(coefs)

x_test = np.linspace(0, 186)
y_pred = predict(x_test[:, None])
plt.scatter(x, y)
plt.plot(x_test, y_pred, c = 'r')
plt.show()

print(predict(186 + 28))
```
