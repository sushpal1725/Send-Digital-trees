```python
from datetime import datetime as dt
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
all_super_tree = pd.read_csv('all_super_tree.csv', delimiter = ' ',
names = ['day','my_date','user_id','super_trees'],
                             usecols =['my_date','super_trees'],parse_dates=[0])

```

```python
super_trees = all_super_tree.groupby('my_date').count()[['super_trees']]
```

```python
super_trees.plot()
```

```python
super_trees_moving_avg= super_trees.rolling(window=10).mean()
```

```python
super_trees_moving_avg.plot()
```

```python
super_trees.describe()
```

```python
dates = super_trees.index.to_julian_date()
```

```python
y = np.asarray(super_trees['super_trees']).reshape(-1,1)
x = np.asarray(dates).reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred= reg.predict(x_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
r2_score(y_test, y_pred)
```

```python
rest_year=pd.date_range(start="2020-06-29",end="2020-12-31")
```

```python
needed_dates= rest_year.to_julian_date()
```

```python
needed_dates = np.asarray(needed_dates).reshape(-1,1)
```

```python
rest_year_predictions = reg.predict(needed_dates).reshape(186)
```

```python
needed_predictions = pd.DataFrame(np.array([rest_year,rest_year_predictions])).T
```

```python
needed_predictions[0] = needed_predictions[0].dt.normalize()
```

```python
needed_predictions.columns = ['future_dates', 'super_tree_predicts']
```

```python
needed_predictions.set_index('future_dates')
```

```python
needed_predictions.plot(x='future_dates', y='super_tree_predicts')
```
