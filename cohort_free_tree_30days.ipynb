```python
import csv
import numpy as np 
```

```python
user_sends={}
with open('free_tree_first_month.csv') as csvfile:
    read = csv.reader(csvfile, delimiter=' ')
    for row in read:
        if row[2] not in user_sends:
            user_sends[row[2]] = []
        if row[1] not in user_sends[row[2]]: 
            user_sends[row[2]].append(row[1])
```

```python
user_sends_list = list(user_sends.values())
```

```python
with open('cohortreport.csv', 'w') as ofile:
    cwrite = csv.writer(ofile, delimiter=';')
    fileds = ['Cohort', 'Days_Passed', 'Number_of_Users']  
    cwrite.writerow(fileds)
    date_list = ['2020-01-10', '2020-01-11', '2020-01-12', '2020-01-13', '2020-01-14', '2020-01-15', '2020-01-16', '2020-01-17', '2020-01-18', '2020-01-19', '2020-01-20', '2020-01-21', '2020-01-22', '2020-01-23', '2020-01-24', '2020-01-25', '2020-01-26', '2020-01-27', '2020-01-28', '2020-01-29', '2020-01-30', '2020-01-31', '2020-02-01', '2020-02-02', '2020-02-03', '2020-02-04', '2020-02-05', '2020-02-06', '2020-02-07', '2020-02-08']
    print("cohort;days_passed;number_of_users")
    for cohort in date_list:
            day_passed = 0
            for given_date in date_list[date_list.index(cohort):date_list.index(cohort)+8]:
                number_of_users = 0
                for users in user_sends_list:
                    if given_date in users and users[0] == cohort:
                        number_of_users = number_of_users + 1
                print(cohort + ";" + str(day_passed) + ";" + str(number_of_users))
                day_passed = day_passed + 1
                cwrite.writerow([cohort, day_passed, number_of_users])
```
